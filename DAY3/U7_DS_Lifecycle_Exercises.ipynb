```python
# ============================================================
# U7 — DS Lifecycle & Data Sources
# Complete Solutions for LAB EXERCISES 1–5
# ============================================================

import pandas as pd
import json
import requests
from collections import Counter

print("Setup complete. pandas", pd.__version__)

# ============================================================
# LAB EXERCISE 1
# Order the Lifecycle
# ============================================================

print("\n" + "="*60)
print("LAB EXERCISE 1 — Order the Lifecycle")
print("="*60)

lifecycle = [
    'Define',
    'Collect',
    'Prepare',
    'Explore',
    'Model',
    'Evaluate',
    'Deploy',
    'Monitor'
]

shuffled = [
    'Model', 'Define', 'Deploy', 'Collect',
    'Monitor', 'Prepare', 'Evaluate', 'Explore'
]

# Sort according to lifecycle order
correct_order = sorted(shuffled, key=lifecycle.index)

print("Correct Order:")
print(correct_order)

print("\nMatches lifecycle?",
      correct_order == lifecycle)

# ============================================================
# LAB EXERCISE 2
# CRISP-DM Grouping
# ============================================================

print("\n" + "="*60)
print("LAB EXERCISE 2 — CRISP-DM Grouping")
print("="*60)

mapping = {
    'Define':   'Business Understanding',
    'Collect':  'Data Understanding',
    'Explore':  'Data Understanding',
    'Prepare':  'Data Preparation',
    'Model':    'Modeling',
    'Evaluate': 'Evaluation',
    'Deploy':   'Deployment',
    'Monitor':  'Deployment',
}

# 1. Invert mapping
grouped = {}

for stage, phase in mapping.items():
    grouped.setdefault(phase, []).append(stage)

# 2. Print phase -> stages
for phase, stages in grouped.items():
    print(f"\n{phase}:")
    print(stages)

# 3. Phase with most stages
largest_phase = max(
    grouped,
    key=lambda p: len(grouped[p])
)

print("\nPhase with most stages:")
print(largest_phase)

# ============================================================
# LAB EXERCISE 3
# Classify Data Types
# ============================================================

print("\n" + "="*60)
print("LAB EXERCISE 3 — Classify Data Types")
print("="*60)

def classify(fmt):
    fmt = fmt.lower()

    if fmt in ('csv', 'sql', 'parquet', 'excel', 'xlsx'):
        return 'structured'

    if fmt in ('json', 'xml', 'yaml', 'log'):
        return 'semi-structured'

    return 'unstructured'

to_label = [
    'parquet',
    'yaml',
    'png',
    'csv',
    'wav',
    'json',
    'txt'
]

labels = [classify(x) for x in to_label]

for item, label in zip(to_label, labels):
    print(f"{item:8s} -> {label}")

# Count categories
counts = Counter(labels)

print("\nCounts:")
print(counts)

# Example answer:
# ML task using unstructured data:
# Image Classification / Sentiment Analysis / Speech Recognition

# ============================================================
# LAB EXERCISE 4
# Round-trip a CSV
# ============================================================

print("\n" + "="*60)
print("LAB EXERCISE 4 — CSV Round Trip")
print("="*60)

# Create DataFrame
products = pd.DataFrame({
    "product": ["Laptop", "Phone", "Tablet"],
    "price": [75000, 30000, 25000],
    "in_stock": [True, True, False]
})

# Write and read back
products.to_csv("products.csv", index=False)

products_read = pd.read_csv("products.csv")

print("Shape:")
print(products_read.shape)

print("\nDtypes:")
print(products_read.dtypes)

print("\nData:")
print(products_read)

# ============================================================
# LAB EXERCISE 5
# Pull Data From API
# ============================================================

print("\n" + "="*60)
print("LAB EXERCISE 5 — REST API")
print("="*60)

POSTS_URL = "https://jsonplaceholder.typicode.com/posts"

try:
    response = requests.get(
        POSTS_URL,
        timeout=10
    )

    response.raise_for_status()

    posts_data = response.json()

    print("Fetched",
          len(posts_data),
          "records from API")

except Exception as e:

    print("Offline mode:",
          type(e).__name__)

    posts_data = [
        {
            "userId": 1,
            "id": 1,
            "title": "Sample Post 1",
            "body": "Example text"
        },
        {
            "userId": 1,
            "id": 2,
            "title": "Sample Post 2",
            "body": "Example text"
        },
        {
            "userId": 2,
            "id": 3,
            "title": "Sample Post 3",
            "body": "Example text"
        }
    ]

# Build DataFrame
posts_df = pd.json_normalize(posts_data)

# Shape + first 3 rows
print("\nShape:")
print(posts_df.shape)

print("\nFirst 3 rows:")
print(posts_df.head(3))

# ============================================================
# ALL U7 LAB EXERCISES COMPLETED
# ============================================================

print("\n" + "="*60)
print("U7 LAB COMPLETED SUCCESSFULLY")
print("="*60)
print("✓ Lifecycle ordering")
print("✓ CRISP-DM grouping")
print("✓ Data type classification")
print("✓ CSV ingestion")
print("✓ REST API ingestion")
```
