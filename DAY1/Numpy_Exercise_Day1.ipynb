```python
# Hands-On #1: Vectors, Dot Products, Norms & Cosine Similarity in NumPy

import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

print("="*60)
print("1. Creating Vectors & Basic Operations")
print("="*60)

u = np.array([2, 3])
v = np.array([4, 1])

print("u =", u)
print("v =", v)
print("u + v =", u + v)
print("3 * u =", 3 * u)

print("\n" + "="*60)
print("2. Dot Product & Angle Between Vectors")
print("="*60)

dot1 = np.dot(u, v)
dot2 = u @ v

cos_theta = np.dot(u, v) / (np.linalg.norm(u) * np.linalg.norm(v))
theta = np.degrees(np.arccos(cos_theta))

print("Dot product using np.dot :", dot1)
print("Dot product using @      :", dot2)
print(f"Cos(theta)               : {cos_theta:.4f}")
print(f"Angle between vectors    : {theta:.2f} degrees")

print("\n" + "="*60)
print("3. Norms")
print("="*60)

w = np.array([3, -4])

print("w =", w)
print("L1 Norm   :", np.linalg.norm(w, 1))
print("L2 Norm   :", np.linalg.norm(w, 2))
print("Linf Norm :", np.linalg.norm(w, np.inf))

print("\n" + "="*60)
print("4. Cosine Similarity Example")
print("="*60)

a = np.array([[1, 2, 3]])
b = np.array([[2, 4, 6]])

print("a =", a)
print("b =", b)
print("Cosine Similarity =", cosine_similarity(a, b)[0][0])

print("\n" + "="*60)
print("5. Exercise Solution: Student Exam Scores")
print("="*60)

# Scores = [Math, Science, English]
student_A = np.array([85, 90, 88])
student_B = np.array([80, 85, 84])
student_C = np.array([95, 70, 75])

students = {
    "Student A": student_A,
    "Student B": student_B,
    "Student C": student_C
}

print("\nStudent Scores:")
for name, scores in students.items():
    print(f"{name}: {scores}")

print("\nL2 Norms (Overall Performance):")
for name, scores in students.items():
    print(f"{name}: {np.linalg.norm(scores):.2f}")

print("\nCosine Similarities:")
print("A vs B =", cosine_similarity([student_A], [student_B])[0][0])
print("A vs C =", cosine_similarity([student_A], [student_C])[0][0])
print("B vs C =", cosine_similarity([student_B], [student_C])[0][0])

print("\nInterpretation:")
print("- Higher L2 norm -> stronger overall performance.")
print("- Cosine similarity close to 1 -> students have similar score patterns.")
print("- Lower cosine similarity -> different strengths across subjects.")
```
