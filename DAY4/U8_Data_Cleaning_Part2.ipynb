```python
# ============================================================
# U8 — Data Cleaning & Preprocessing (Part 2)
# Complete Solutions for LAB EXERCISES 1–5
# ============================================================

import numpy as np
import pandas as pd

pd.set_option('display.max_columns', None)

print("Setup complete. pandas", pd.__version__)

# ============================================================
# DATASET
# ============================================================

df = pd.DataFrame({
    'city':   ['pune', 'delhi', 'mumbai', 'pune',
               'delhi', 'mumbai', 'pune', 'delhi'],
    'size':   ['small', 'large', 'medium', 'medium',
               'small', 'large', 'large', 'small'],
    'age':    [25, 41, 33, 29, 52, 38, 46, 22],
    'income': [38000, 92000, 55000, 47000,
               120000, 76000, 88000, 41000],
    'bought': [0, 1, 0, 0, 1, 1, 1, 0]
})

# ============================================================
# LAB EXERCISE 1
# Encoding Categorical Variables
# ============================================================

print("\n" + "="*60)
print("LAB EXERCISE 1 — Encoding")
print("="*60)

# 1. One-hot encode city in whole DataFrame
df_ohe = pd.get_dummies(df, columns=['city'], dtype=int)

print("One-Hot Encoded DataFrame:")
print(df_ohe.head())

# 2. Ordinal encoding for size

size_order = {
    'small': 0,
    'medium': 1,
    'large': 2
}

df['size_code'] = df['size'].map(size_order)

print("\nSize Encoding:")
print(df[['size', 'size_code']])

# 3. Explanation:
# One-hot is wrong for size because size has an order
# (small < medium < large).
#
# Ordinal is wrong for city because city names have
# no natural ranking and assigning numbers would imply
# a false order.

# ============================================================
# LAB EXERCISE 2
# Feature Scaling
# ============================================================

print("\n" + "="*60)
print("LAB EXERCISE 2 — Feature Scaling")
print("="*60)

from sklearn.preprocessing import StandardScaler, MinMaxScaler

income = df[['income']]

# 1. Standardization
std_scaler = StandardScaler()

income_std = std_scaler.fit_transform(income)

print("Standardized Mean:",
      round(income_std.mean(), 6))

print("Standardized Std :",
      round(income_std.std(), 6))

# 2. Min-Max Scaling

mm_scaler = MinMaxScaler()

income_mm = mm_scaler.fit_transform(income)

print("\nMin-Max Min:",
      income_mm.min())

print("Min-Max Max:",
      income_mm.max())

# 3. Examples:
# Needs scaling:
# - KNN
# - SVM
# - Logistic Regression
#
# Doesn't need scaling:
# - Decision Tree
# - Random Forest

# ============================================================
# LAB EXERCISE 3
# Data Leakage
# ============================================================

print("\n" + "="*60)
print("LAB EXERCISE 3 — Data Leakage")
print("="*60)

from sklearn.model_selection import train_test_split

X = df[['age', 'income']]
y = df['bought']

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.25,
    random_state=0
)

# Wrong
wrong_mean = StandardScaler().fit(X).mean_

# Correct
right_mean = StandardScaler().fit(X_train).mean_

print("Fit-on-all mean :",
      np.round(wrong_mean, 1))

print("Fit-on-train mean:",
      np.round(right_mean, 1))

# Explanation:
# Fitting on all data lets information from the
# test set influence preprocessing statistics.
# This causes overly optimistic evaluation results.

# ============================================================
# LAB EXERCISE 4
# Feature Engineering
# ============================================================

print("\n" + "="*60)
print("LAB EXERCISE 4 — Feature Engineering")
print("="*60)

ex = df.copy()

# 1. High earner flag

median_income = ex['income'].median()

ex['high_earner'] = ex['income'] > median_income

# 2. Income buckets

ex['income_group'] = pd.cut(
    ex['income'],
    bins=3,
    labels=['low', 'mid', 'high']
)

# 3. Display

print(ex[['income',
          'high_earner',
          'income_group']])

# ============================================================
# LAB EXERCISE 5
# Pipeline
# ============================================================

print("\n" + "="*60)
print("LAB EXERCISE 5 — ML Pipeline")
print("="*60)

from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression

Xn = df[['age', 'income']]
yn = df['bought']

# 1. Pipeline

pipe = Pipeline([
    ('scaler', MinMaxScaler()),
    ('model', LogisticRegression())
])

# 2. Train/Test Split

X_train, X_test, y_train, y_test = train_test_split(
    Xn,
    yn,
    test_size=0.25,
    random_state=0
)

pipe.fit(X_train, y_train)

# 3. Accuracy

acc = pipe.score(X_test, y_test)

print("Test Accuracy:",
      round(acc, 3))

# ============================================================
# BONUS: Compare Original vs Scaled Features
# ============================================================

print("\n" + "="*60)
print("BONUS — Scaling Example")
print("="*60)

scaled_df = pd.DataFrame(
    StandardScaler().fit_transform(
        df[['age', 'income']]
    ),
    columns=['age_scaled', 'income_scaled']
)

print(scaled_df.head())

print("\n" + "="*60)
print("U8 PART-2 LAB COMPLETED")
print("="*60)
print("✓ One-Hot Encoding")
print("✓ Ordinal Encoding")
print("✓ Standardization")
print("✓ Min-Max Scaling")
print("✓ Data Leakage Prevention")
print("✓ Feature Engineering")
print("✓ Scikit-Learn Pipelines")
print("✓ Model-Ready Data Preparation")
```
