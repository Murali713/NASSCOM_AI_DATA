```python
# ============================================================
# U5 — Calculus for ML: LAB EXERCISE SOLUTIONS (Single Cell)
# ============================================================

import numpy as np
import sympy as sp

x, y = sp.symbols('x y')

print("="*60)
print("LAB EXERCISE 1 — Numerical vs Symbolic Derivative")
print("="*60)

def g(x):
    return x**3 + 2*x

h = 1e-6
num_deriv = (g(2 + h) - g(2)) / h

expr = x**3 + 2*x
sym_deriv = sp.diff(expr, x)
sym_value = sym_deriv.subs(x, 2)

print("Numerical derivative at x=2:", num_deriv)
print("Symbolic derivative:", sym_deriv)
print("Symbolic derivative at x=2:", sym_value)

print("\n" + "="*60)
print("LAB EXERCISE 2 — Gradient")
print("="*60)

h2 = x**2 * y + sp.sin(y)

dh_dx = sp.diff(h2, x)
dh_dy = sp.diff(h2, y)

grad_h2 = [dh_dx, dh_dy]
grad_eval = [g.subs({x: 2, y: 0}) for g in grad_h2]

print("dh/dx =", dh_dx)
print("dh/dy =", dh_dy)
print("Gradient =", grad_h2)
print("Gradient at (2,0) =", grad_eval)

print("\n" + "="*60)
print("LAB EXERCISE 3 — Chain Rule")
print("="*60)

by_hand = sp.exp(x**2 + 1) * (2*x)
by_sympy = sp.diff(sp.exp(x**2 + 1), x)

print("By hand :", by_hand)
print("By SymPy:", by_sympy)
print("Match?  ", sp.simplify(by_hand - by_sympy) == 0)

print("\n" + "="*60)
print("LAB EXERCISE 4 — Backpropagation")
print("="*60)

np.random.seed(42)

Xb = np.random.randn(6, 4)
Yb = np.random.randn(6, 1)

Wa = np.random.randn(4, 8) * 0.1
Wb = np.random.randn(8, 1) * 0.1

# Forward Pass
z1 = Xb @ Wa
h = np.maximum(0, z1)

y_hat = h @ Wb
loss = ((y_hat - Yb) ** 2).mean()

print("Initial Loss:", round(loss, 6))

# Backward Pass
dy = 2 * (y_hat - Yb) / Yb.size

dWb = h.T @ dy

dh = dy @ Wb.T

dz1 = dh * (z1 > 0)

dWa = Xb.T @ dz1

print("dWa shape:", dWa.shape)
print("dWb shape:", dWb.shape)

# Gradient Descent Step
lr = 0.05

Wa -= lr * dWa
Wb -= lr * dWb

z1_new = Xb @ Wa
h_new = np.maximum(0, z1_new)

y_hat_new = h_new @ Wb

loss_new = ((y_hat_new - Yb) ** 2).mean()

print("Loss before:", round(loss, 6))
print("Loss after :", round(loss_new, 6))

print("\n" + "="*60)
print("LAB EXERCISE 5 — Hessian & Gradient Descent")
print("="*60)

# Hessian
f = x**4 + y**2

H = sp.hessian(f, (x, y))

print("Hessian of x^4 + y^2:")
sp.pprint(H)

# Gradient Descent
xv = 0.0
lr = 0.1

for step in range(20):
    grad = 2 * (xv - 7)
    xv = xv - lr * grad

print("\nFinal x after Gradient Descent:", round(xv, 6))

print("\n" + "="*60)
print("SUMMARY")
print("="*60)
print("✓ Numerical Derivatives")
print("✓ Symbolic Derivatives")
print("✓ Partial Derivatives")
print("✓ Gradients")
print("✓ Chain Rule")
print("✓ Backpropagation")
print("✓ Hessian Matrix")
print("✓ Gradient Descent")
```
