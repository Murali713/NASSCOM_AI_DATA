```python
# ============================================================
# U6 — Probability & Statistics (Part 1)
# Complete Solutions for LAB EXERCISES 1–5
# ============================================================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")
np.random.seed(42)

# Load Dataset
df = sns.load_dataset("tips")

print("="*60)
print("LAB EXERCISE 1 — Classify Columns")
print("="*60)

# 1. Numerical vs Categorical
numerical = df.select_dtypes(include="number").columns.tolist()
categorical = df.select_dtypes(exclude="number").columns.tolist()

print("Numerical Columns:")
print(numerical)

print("\nCategorical Columns:")
print(categorical)

# 2. Mean of tip
print("\nMean Tip:")
print(round(df["tip"].mean(), 2))

# 3. Value counts of sex
print("\nValue Counts of Sex:")
print(df["sex"].value_counts())

# ============================================================
# LAB EXERCISE 2
# Central Tendency
# ============================================================

print("\n" + "="*60)
print("LAB EXERCISE 2 — Central Tendency")
print("="*60)

tip = df["tip"]

# Mean, Median, Mode
print("Mean  :", round(tip.mean(), 2))
print("Median:", round(tip.median(), 2))
print("Mode  :", round(tip.mode()[0], 2))

# Right Skew?
print("\nRight Skewed?")
print(tip.mean() > tip.median())

# Histogram
plt.figure(figsize=(7,4))
sns.histplot(tip, kde=True)

plt.axvline(
    tip.mean(),
    color='red',
    label='Mean'
)

plt.axvline(
    tip.median(),
    color='green',
    label='Median'
)

plt.legend()
plt.title("Tip Distribution")
plt.show()

# ============================================================
# LAB EXERCISE 3
# Measures of Spread
# ============================================================

print("\n" + "="*60)
print("LAB EXERCISE 3 — Spread")
print("="*60)

# Range
tip_range = tip.max() - tip.min()

# Variance
tip_var = tip.var()

# Standard Deviation
tip_std = tip.std()

print("Range :", round(tip_range, 2))
print("Variance :", round(tip_var, 2))
print("Std Dev :", round(tip_std, 2))

# Quartiles and IQR
Q1 = tip.quantile(0.25)
Q3 = tip.quantile(0.75)
IQR = Q3 - Q1

print("\nQ1 :", round(Q1, 2))
print("Q3 :", round(Q3, 2))
print("IQR:", round(IQR, 2))

# Most robust measure of spread:
# IQR because it uses the middle 50% of data
# and is less affected by outliers.

# ============================================================
# LAB EXERCISE 4
# Five Number Summary & Box Plot
# ============================================================

print("\n" + "="*60)
print("LAB EXERCISE 4 — Shape Analysis")
print("="*60)

five_num = tip.describe()[['min', '25%', '50%', '75%', 'max']]

print("Five Number Summary:")
print(five_num)

# Skewness
skewness = tip.skew()

print("\nSkewness:")
print(round(skewness, 3))

# Positive skew => right tail
# Negative skew => left tail

plt.figure(figsize=(7,3))
sns.boxplot(x=tip)

plt.title("Box Plot of Tip")
plt.show()

# ============================================================
# LAB EXERCISE 5
# Correlation & Dataset Summary
# ============================================================

print("\n" + "="*60)
print("LAB EXERCISE 5 — Correlation & Summary")
print("="*60)

# Correlation Matrix
corr = df.corr(numeric_only=True)

print("Correlation Matrix:")
print(corr.round(2))

# Heatmap
plt.figure(figsize=(6,4))

sns.heatmap(
    corr,
    annot=True,
    cmap="Blues",
    vmin=-1,
    vmax=1
)

plt.title("Correlation Heatmap")
plt.show()

# Dataset Summary
summary = df.describe()

print("\nDataset Summary:")
print(summary)

# ------------------------------------------------
# Insights
# ------------------------------------------------
# Insight 1:
# total_bill and tip show strong positive correlation.
#
# Insight 2:
# total_bill has the largest spread (highest std deviation).
#
# Insight 3:
# tip distribution is positively skewed
# (mean > median and positive skewness).
#
# ------------------------------------------------

print("\nAll U6 Lab Exercises Completed Successfully!")
```
