```python
import pandas as pd
import numpy as np

print("--- Intermediate Pandas Exercise Solution ---")

# ==================================================
# Task 1: Basic Pandas Functions and Converting Arrays
# ==================================================
print("\n--- Task 1 ---")

# 1.1 Series
s1_1 = pd.Series([10, 20, 30, 40, 50])

# 1.2 NumPy Array
np_array1_2 = np.random.randint(1, 101, size=(3,4))

# 1.3 DataFrame
df1_3 = pd.DataFrame(np_array1_2, columns=['A', 'B', 'C', 'D'])

print("Series:")
print(s1_1)

print("\nNumPy Array:")
print(np_array1_2)

print("\nDataFrame:")
print(df1_3)

# ==================================================
# Task 2: Synthetic Data Generation
# ==================================================
print("\n--- Task 2 ---")

np.random.seed(42)

df_sales = pd.DataFrame({
    'Date': pd.date_range('2023-01-01', periods=100),
    'Region': np.random.choice(['East', 'West', 'North', 'South'], 100),
    'Product': np.random.choice(['Laptop', 'Mouse', 'Keyboard', 'Monitor'], 100),
    'Sales': np.random.randint(100, 1001, 100),
    'Quantity': np.random.randint(1, 11, 100)
})

# Introduce NaN values
sales_nan_idx = np.random.choice(df_sales.index, 5, replace=False)
qty_nan_idx = np.random.choice(df_sales.index, 3, replace=False)

df_sales.loc[sales_nan_idx, 'Sales'] = np.nan
df_sales.loc[qty_nan_idx, 'Quantity'] = np.nan

print(df_sales.head())

# ==================================================
# Task 3: Indexing and Slicing
# ==================================================
print("\n--- Task 3 ---")

# 3.1
df3_1 = df_sales[['Region', 'Sales']]

# 3.2
df3_2 = df_sales[df_sales['Product'] == 'Laptop']

# 3.3
df3_3 = df_sales[
    (df_sales['Region'] == 'East') &
    (df_sales['Sales'] > 500)
]

# 3.4
monitor_idx = df_sales[df_sales['Product'] == 'Monitor'].index[0]
val3_4 = df_sales.loc[monitor_idx, 'Sales']

print("Region & Sales:")
print(df3_1.head())

print("\nLaptop Records:")
print(df3_2.head())

print("\nEast Region with Sales > 500:")
print(df3_3.head())

print("\nFirst Monitor Sales Value:", val3_4)

# ==================================================
# Task 4: Data Cleaning
# ==================================================
print("\n--- Task 4 ---")

# 4.1
missing_counts4_1 = df_sales.isnull().sum()

# 4.2
df4_2_filled = df_sales.copy()
sales_mean = df4_2_filled['Sales'].mean()
df4_2_filled['Sales'] = df4_2_filled['Sales'].fillna(sales_mean)

# 4.3
df4_3_dropped = df_sales.dropna()

print("Missing Values:")
print(missing_counts4_1)

print("\nFilled Sales NaNs with Mean:")
print(df4_2_filled.head())

print("\nRows After Dropping NaNs:")
print(df4_3_dropped.head())

# ==================================================
# Task 5: Data Manipulation
# ==================================================
print("\n--- Task 5 ---")

# 5.1
df_sales['Total_Revenue'] = (
    df_sales['Sales'] * df_sales['Quantity']
)

# 5.2
df5_2_sorted = df_sales.sort_values(
    by=['Date', 'Sales'],
    ascending=[True, False]
)

# 5.3
df5_3_grouped_region = (
    df_sales.groupby('Region')[['Sales', 'Quantity']]
    .sum()
)

# 5.4
df5_4_grouped_product = (
    df_sales.groupby('Product')['Sales']
    .mean()
)

print("\nData with Total Revenue:")
print(df_sales.head())

print("\nSorted Data:")
print(df5_2_sorted.head())

print("\nRegion-wise Total Sales and Quantity:")
print(df5_3_grouped_region)

print("\nAverage Sales by Product:")
print(df5_4_grouped_product)

print("\nExercise Completed Successfully!")
```
