```
import os
import datetime
import random

print("--- Intermediate Python Core Exercise ---")
print("This exercise covers Python fundamentals, data structures, functions,")
print("core libraries (os, datetime, random), file handling, and exception handling.")
print("Estimated completion time: 3-4 hours.")
print("------------------------------------------")

# ============================================================
# Global Data (Simulated Sensor Environment)
# ============================================================

DEVICE_TYPES = ['Temperature_Sensor', 'Humidity_Sensor', 'Pressure_Sensor', 'Light_Sensor']
LOCATIONS = ['Lab1', 'Lab2', 'OfficeA', 'OfficeB', 'Warehouse']

READING_RANGES = {
    'Temperature_Sensor': (18.0, 30.0),
    'Humidity_Sensor': (30.0, 80.0),
    'Pressure_Sensor': (980.0, 1030.0),
    'Light_Sensor': (100, 1000)
}

# ============================================================
# Task 1: Data Generation and Basic Structures
# ============================================================

print("\n--- Task 1: Data Generation and Basic Structures ---")

# ------------------------------------------------------------
# 1.1 Generate 100 simulated sensor readings
# Each reading:
# (timestamp_str, device_id, device_type, location, value)
# ------------------------------------------------------------

raw_sensor_data = []

base_time = datetime.datetime.now()

for i in range(100):

    device_type = random.choice(DEVICE_TYPES)
    location = random.choice(LOCATIONS)

    min_val, max_val = READING_RANGES[device_type]

    value = round(random.uniform(min_val, max_val), 2)

    timestamp = (
        base_time + datetime.timedelta(minutes=i)
    ).strftime('%Y-%m-%d %H:%M:%S')

    device_id = f"DEV{i+1:03d}"

    raw_sensor_data.append(
        (timestamp, device_id, device_type, location, value)
    )
    print('Raw data','\n', raw_sensor_data)
print(f"Generated {len(raw_sensor_data)} sensor readings.")

# ------------------------------------------------------------
# 1.2 Extract only temperature sensor readings
# Convert each reading into a dictionary
# ------------------------------------------------------------

temperature_readings = []

for reading in raw_sensor_data:

    if reading[2] == 'Temperature_Sensor':

        temperature_readings.append({
            'timestamp': reading[0],
            'device_id': reading[1],
            'location': reading[3],
            'temperature_celsius': reading[4]
        })

print(f"Temperature readings found: {len(temperature_readings)}")

# ------------------------------------------------------------
# 1.3 Create a set of unique locations
# ------------------------------------------------------------

unique_locations = {reading[3] for reading in raw_sensor_data}

print("Unique Locations:", unique_locations)

# ------------------------------------------------------------
# 1.4 Create device statistics dictionary
# Format:
# {
#   device_type : (min_value, max_value)
# }
# ------------------------------------------------------------

device_stats = {}

for device_type in DEVICE_TYPES:
    device_stats[device_type] = (float('inf'), float('-inf'))

for reading in raw_sensor_data:

    dtype = reading[2]
    value = reading[4]

    current_min, current_max = device_stats[dtype]

    device_stats[dtype] = (
        min(current_min, value),
        max(current_max, value)
    )

print("\nDevice Statistics:")
for device, stats in device_stats.items():
    print(f"{device}: Min={stats[0]}, Max={stats[1]}")

# ============================================================
# 1.5 Advanced Indexing and Slicing
# ============================================================

print("\n--- Task 1.5: Advanced Indexing and Slicing ---")

sample_string = "PythonProgrammingIsFun"

# ------------------------------------------------------------
# 1.5.1 List Slicing
# ------------------------------------------------------------

# Elements 10 through 20 inclusive
sliced_readings_1 = raw_sensor_data[9:20]

# Every 5th reading
sliced_readings_2 = raw_sensor_data[::5]

print("Length of sliced_readings_1:", len(sliced_readings_1))
print("Length of sliced_readings_2:", len(sliced_readings_2))

# ------------------------------------------------------------
# 1.5.2 String Slicing
# ------------------------------------------------------------

substring_1 = sample_string[6:17]     # Programming
substring_2 = sample_string[5::-1]    # nohtyP

print("Substring 1:", substring_1)
print("Substring 2:", substring_2)

# ------------------------------------------------------------
# 1.5.3 Mixed Slicing
# ------------------------------------------------------------

fifth_reading_device_type = raw_sensor_data[4][2]

device_type_suffix = fifth_reading_device_type[-6:]

print("5th Reading Device Type:", fifth_reading_device_type)
print("Suffix:", device_type_suffix)

# ============================================================
# Task 2: Function Definitions and Usage
# ============================================================

print("\n--- Task 2: Function Definitions and Usage ---")

# ------------------------------------------------------------
# 2.1 Get all readings for a given location
# ------------------------------------------------------------

def get_readings_by_location(readings_list, target_location):
    """
    Return all sensor readings for a specific location.
    """
    return [
        reading
        for reading in readings_list
        if reading[3] == target_location
    ]


# ------------------------------------------------------------
# 2.2 Calculate average temperature
# ------------------------------------------------------------

def calculate_average_temperature(temp_data_list):
    """
    Calculate average temperature.
    Returns 0.0 if list is empty.
    """

    if not temp_data_list:
        return 0.0

    total = sum(
        reading['temperature_celsius']
        for reading in temp_data_list
    )

    return total / len(temp_data_list)


# ------------------------------------------------------------
# 2.3 Generate formatted report line
# ------------------------------------------------------------

def generate_report_line(reading_dict):
    """
    Generate a human-readable report string.
    """

    return (
        f"[{reading_dict['timestamp']}] "
        f"Device {reading_dict['device_id']} "
        f"at {reading_dict['location']}: "
        f"{reading_dict['temperature_celsius']}°C"
    )


# Test functions

avg_temp = calculate_average_temperature(temperature_readings)

print(f"\nAverage Temperature: {avg_temp:.2f}°C")

lab1_readings = get_readings_by_location(raw_sensor_data, "Lab1")

print(f"Readings from Lab1: {len(lab1_readings)}")

# ============================================================
# Task 3: Core Python Libraries
# ============================================================

print("\n--- Task 3: Core Library Usage ---")

# ------------------------------------------------------------
# 3.1 Create directory if it doesn't exist
# ------------------------------------------------------------

reports_dir = "sensor_reports"

if not os.path.exists(reports_dir):
    os.makedirs(reports_dir)
    print(f"Directory '{reports_dir}' created.")
else:
    print(f"Directory '{reports_dir}' already exists.")

# ------------------------------------------------------------
# 3.2 Calculate duration between first and last timestamp
# ------------------------------------------------------------

if len(raw_sensor_data) > 1:

    first_timestamp_str = raw_sensor_data[0][0]
    last_timestamp_str = raw_sensor_data[-1][0]

    date_format = '%Y-%m-%d %H:%M:%S'

    first_dt = datetime.datetime.strptime(
        first_timestamp_str,
        date_format
    )

    last_dt = datetime.datetime.strptime(
        last_timestamp_str,
        date_format
    )

    duration = last_dt - first_dt

    total_seconds = int(duration.total_seconds())

    days = duration.days
    hours = (total_seconds // 3600) % 24
    minutes = (total_seconds % 3600) // 60

    print(
        f"Duration of data collection: "
        f"{days} days, {hours} hours, {minutes} minutes."
    )

# ------------------------------------------------------------
# 3.3 Generate 5 random integers between 100 and 200
# ------------------------------------------------------------

random_values = [random.randint(100, 200) for _ in range(5)]

print("Random Values:", random_values)

# ============================================================
# Task 4: File Handling
# ============================================================

print("\n--- Task 4: File I/O ---")

# ------------------------------------------------------------
# 4.1 Write temperature readings to file
# ------------------------------------------------------------

report_filename = os.path.join(
    reports_dir,
    "temperature_report.txt"
)

try:

    with open(report_filename, "w") as file:

        for reading in temperature_readings:

            line = generate_report_line(reading)

            file.write(line + "\n")

    print(f"Temperature report written to '{report_filename}'.")

except IOError as e:

    print(f"Error writing file: {e}")

# ------------------------------------------------------------
# 4.2 Read file content back into a list
# ------------------------------------------------------------

read_report_lines = []

try:

    with open(report_filename, "r") as file:

        read_report_lines = file.readlines()

    print("\nFirst 3 lines from report:")

    for line in read_report_lines[:3]:
        print(line.strip())

except IOError as e:

    print(f"Error reading file: {e}")

# ============================================================
# Exercise Complete
# ============================================================

print("\n--- Exercise Complete! ---")
print("Great job completing the intermediate Python exercise.")
print("Review your code and outputs to ensure correctness.")
print("Try extending the project with user input, CSV export,")
print("or additional sensor analytics for more practice.")
```
